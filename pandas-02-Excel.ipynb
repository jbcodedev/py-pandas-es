## <font color=green> **02 - Leyendo archivos Excel**

### <font color=green> **Importando archivos Excel**

```python
# Importando Pandas

import pandas as pd
```

```python
# Asignando a una variable la ruta del archivo Excel

archivo = 'data\\emisiones_CO2.xlsx'
```

```python
# Asignando a un dataframe la informacion leida por la funcion Pandas read_excel

datos_co2 = pd.read_excel(archivo)
```

```python
# Leyendo las primeras lineas del dataframe creado

datos_co2.head()
```

```python
# Visualizando los nombres de las hojas del archivo Excel

pd.ExcelFile(archivo).sheet_names
```

### <font color=green> **Parámetros de la funcion read_excel**

Documentación: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html

```python
# Usando el parametro "sheet_name", para elegir la hoja que se desea leer
# Visualizando las primeras 5 filas del dataframe

percapita = pd.read_excel(archivo, sheet_name='emisiones_percapita')
percapita.sample(5)
```

```python
# Usando el parametro "sheet_name" para elegir la otra hoja de calculo
# Visualizando las ultimas filas del dataframe

fuentes = pd.read_excel(archivo, sheet_name='fuentes')
fuentes.tail()
```

```python
# Seleccionando un intervalo de columnas con "usecols" en la hoja seleccionada

intervalo = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D')
intervalo
```

```python
# Adicionando el parametro "nrows" para elegir las filas deseadas

intervalo_2 = pd.read_excel(archivo, sheet_name='emisiones_C02', usecols='A:D', nrows=10)
intervalo_2
```
